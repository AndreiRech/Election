```python
import pandas as pd
```

```python
despesas = pd.read_csv('Estrela\\Fatos_Despesas.csv', delimiter=',', encoding='utf-8')
despesas.head()
```

```python
despesas['VR_DESPESA_CONTRATADA'] = despesas['VR_DESPESA_CONTRATADA'].str.replace(',', '.').astype(float)

despesas_final = despesas.groupby('SQ_CANDIDATO')['VR_DESPESA_CONTRATADA'].sum()

print(despesas_final)
```

```python
votos = pd.read_csv('Estrela\\Fatos_Votos.csv', delimiter=',', encoding='utf-8')
votos.head()
```

```python
votos_final = votos.groupby('SQ_CANDIDATO')['QT_VOTOS'].sum().drop(-1, errors='ignore')

print(votos_final)
```

```python
resultado_final = pd.merge(votos_final, despesas_final, on='SQ_CANDIDATO', how='inner')

resultado_final
```

```python
resultado_final = resultado_final.reset_index()

resultado_final.rename(columns={'index': 'SQ_CANDIDATO'}, inplace=True)

resultado_final = resultado_final.loc[:, ~resultado_final.columns.duplicated()]

resultado_final
```

```python
candidatos_original = pd.read_csv('Resultados\\Candidatos.csv', delimiter=',', encoding='utf-8')

candidatos_original = candidatos_original.drop(columns=['ANO_ELEICAO', 'DS_CARGO', 'NM_CANDIDATO', 'NR_PARTIDO', 'PASSOU_TURNO', 'SG_PARTIDO'])

candidatos_original
```

```python
data = pd.merge(resultado_final, candidatos_original, on='SQ_CANDIDATO', how='inner')

data
```

```python
data = data.drop(columns=['SQ_CANDIDATO', 'SG_UF'])
```

```python
data.to_csv('Modelo\\Data.csv', index=False)
```
